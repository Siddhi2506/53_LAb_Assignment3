# Task 1: Environment Setup

```python
# Install necessary libraries
!pip install ultralytics --quiet
import torch
from ultralytics import YOLO

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")
```

#Task 2: Dataset Preparation

```python
# Mount Google Drive if dataset is on drive (optional)
#from google.colab import drive
#drive.mount('/content/drive')

# Upload and unzip dataset
from zipfile import ZipFile
import os

uploaded_path = '/content/3d_model_car.v1i.yolov8.zip'  # Update with your uploaded file path
extracted_path = '/content/3d_model_car'

# Unzip
with ZipFile(uploaded_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)

# Verify
os.listdir(extracted_path)
```

# Task 3: Training YOLOv8 Model

```python
# Load Pretrained YOLOv8 model
model = YOLO('yolov8n.pt')  # YOLOv8 nano version for faster training

# Train on your custom dataset
model.train(data='/content/3d_model_car/data.yaml', epochs=10, imgsz=640, device=device)
```

# Task 4: Inference and Evaluation

```python
import os
for root, dirs, files in os.walk('/content/runs/detect'):
    for file in files:
        if file.endswith('.pt'):  # Look for files ending with .pt (model weights)
            print(os.path.join(root, file))
```

```python
# Validate model
metrics = model.val()

# Print basic metrics
print("\nValidation Results:")
print(f"mAP50-95: {metrics.box.map:.4f}")
print(f"Precision: {metrics.box.p}")  # Print the entire array
print(f"Recall: {metrics.box.r}")  # Print the entire array

# OR, to print the average precision/recall:

#print(f"Precision: {metrics.box.p.mean():.4f}")  # Print average precision
#print(f"Recall: {metrics.box.r.mean():.4f}")  # Print average recall


# F1 Score Calculation
# Modified to use metrics.box.p and metrics.box.r, using mean for overall score
f1_score = 2 * (metrics.box.p.mean() * metrics.box.r.mean()) / (metrics.box.p.mean() + metrics.box.r.mean() + 1e-6)
print(f"F1 Score: {f1_score:.4f}")

# Test predictions
results = model.predict(source='/content/3d_model_car/test/images', save=True)  # Or on any image
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from ultralytics import YOLO

# Locate the Actual Model Path:
print(os.listdir('/content/runs/detect')) # List the contents of the 'detect' directory
print(os.listdir('/content/runs/detect/train2')) # List the contents of the weight directory if you have train54

# Load the trained model
model = YOLO('/content/runs/detect/train5/weights/best.pt')  # Update path to your best.pt

# Path to your test image
image_path = '/content/3d_model_car/test/images/_bus_30_s_50_24_jpg.rf.27faae1b5ce9641c8f844fc8fdb3a62e.jpg'  # Update with your test image

# Perform prediction
results = model.predict(source=image_path, save=False)  # save=False to not save to disk

# Get the first result (assuming you only have one image)
result = results[0]

# Load the image
img = mpimg.imread(image_path)

# Plot the image with bounding boxes
plt.figure(figsize=(10, 10))
plt.imshow(img)
ax = plt.gca()

# Plot bounding boxes and labels
for *xyxy, conf, cls in result.boxes.data:
    x1, y1, x2, y2 = map(int, xyxy)
    label = f'{model.names[int(cls)]} {conf:.2f}'
    ax.add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, fill=False, color='red', linewidth=2))
    ax.text(x1, y1, label, color='white', fontsize=12, bbox=dict(facecolor='red', alpha=0.5))

plt.axis('off')
plt.title('Car Detection Results')
plt.show()
```
